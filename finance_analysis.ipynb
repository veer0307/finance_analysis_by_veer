```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import matplotlib.pyplot as plt
sns.set_style("darkgrid")
```

```python
df=pd.read_csv("/content/Finance_data.csv")
df
```

```python
df.drop(["Mutual_Funds",	"Equity_Market",	"Debentures",	"Government_Bonds",	"Fixed_Deposits",	"PPF",	"Gold"],axis=1,inplace=True)
df
```

```python
sns.countplot(df["gender"],linewidth=3,palette="Set1",edgecolor="black")
plt.show()
```

```python
plt.figure(figsize=(12,12))
sns.countplot(df["age"],linewidth=3,palette="Set3",edgecolor="black")
plt.show()
```

```python
sns.countplot(x=df["Factor"],linewidth=3,palette="coolwarm",edgecolor="black",legend=False)
plt.show()
```

```python
sns.countplot(x=df["gender"],linewidth=3,palette="coolwarm",edgecolor="black",hue=df["Factor"],legend=False)
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.pointplot(x="Purpose",y="age",data=df,linestyles="--",capsize=0.3,color="red")
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(hue=df['gender'],x=df['Duration'],palette='viridis',linewidth=2,edgecolor='black')
plt.title('DURATION')
plt.subplot(1, 2, 2)
sns.countplot(hue=df['gender'],x=df['Invest_Monitor'],palette='seismic',linewidth=2,edgecolor='black')
plt.title('INVESTMENT MONITORING')
plt.show()
```

```python
plt.figure(figsize=(18,6))
plt.subplot(1, 2, 1)
sns.countplot(x=df['gender'],hue=df['Investment_Avenues'],palette='summer',linewidth=3,edgecolor='white')
plt.title('INVESTMENT AVENUE')
plt.subplot(1, 2, 2)
sns.countplot(x=df['gender'],hue=df['Stock_Marktet'],palette='hot',linewidth=3,edgecolor='white')
plt.title('STOCK MARKET')
plt.show()

```
